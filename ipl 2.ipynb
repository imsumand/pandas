#  Q1.- Make a pivot table of top 10 bowlers(in terms of wickets) with over(1-20) showing no. of wickets.  [Dificult-]

```python
m1=delivery['dismissal_kind']=='bowled'
m2=delivery['dismissal_kind']=='caught'
m3=delivery['dismissal_kind']=='caught and bowled'
m4=delivery['dismissal_kind']=='hit wicket'
m5=delivery['dismissal_kind']=='lbw'
m6=delivery['dismissal_kind']=='stumped'
m=m1|m2|m3|m4|m5|m6
wick=delivery[m]
a=wick.groupby('bowler').count()['player_dismissed'].sort_values(ascending=False).head(10)
b=a.reset_index()
c=b['bowler'].reset_index()
d=c.merge(wick)
```

```python
a=d.pivot_table(index='over',columns='bowler',values='player_dismissed',aggfunc='count')
```

# Q2.- Make a function for batting profile of any batsman:-                                              [Medium+]



# Ex-

V Kohli:

Innings:

Total Runs:

No. of MOM(Man of the match):

No. of boundries:

```python
def definer(name):
    x=delivery['batsman']==name
    ok=delivery[x]
    runs=ok['batsman_runs'].sum()
    innings=ok.drop_duplicates(subset=['match_id']).count()['inning']
    motm=match[match['player_of_match']==name].count()['player_of_match']
    fours=ok[ok['batsman_runs']==4].count()['ball']
    sixes=ok[ok['batsman_runs']==6].count()['ball']
    boundaries=fours+sixes
    print(name)
    print ("Innings:",innings)
    print ("Total Runs:",runs)
    print("Man of the Match awards:",motm)
    print ("NO. of boundaries:",boundaries)
```

```python
definer("RG Sharma")
```

# Q3.- Write a function to return top-4 batsman at a venue. Take venue as input.

```python
def orangecaps(venue):
    var2=delivery.merge(match,left_on='match_id',right_on='id')
    var3=var2['venue']==venue
    top=var2[var3].groupby('batsman').count()['batsman_runs'].sort_values(ascending=False).head(4)
    return top
```

```python
orangecaps("Wankhede Stadium")
```
